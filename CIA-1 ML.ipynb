```python
import numpy as np
```

# 1.Generating 1000 random numbers

```python
np.random.seed(5)
x = np.random.randn(1000)

y = np.random.randn(1000) 
```

```python
plt.scatter(x,y)
```

# 2.Simple Linear Regression

```python
x1=x-x.mean()
```

```python
y1=y-y.mean()
```

```python
num=x1*y1 #NUMERATOR
```

```python
den=x1*x1 #DENOMINATOR
```

```python
m=num.sum()/den.sum() #SLOPE
```

```python
print("Slope is : ",m)
```

```python
c=y.mean()-(m*x.mean())
print('C Value: ',c) #INTERCEPT
```

```python
ycap=m*x+c #PREDICTED VALUE
```

```python
e=y-ycap
mse=(e*e).sum()
print('MSE:',mse) #MEAN SQUARED ERROR
```

# 3.Gradient Descent

```python
p = 0
q = 0

l= 0.01 #LEARNING RATE

it = 1000 #number of iterations

for i in range(it):
    y_pred = q * x + p
    err = y - y_pred
    p_grad = -(1/len(x)) * sum(err)
    q_grad = -(1/len(x)) * sum(err * x)
    p = p - l * p_grad
    q = q - l * q_grad

print("Intercept p:", p)#INTERCEPT
print("Slope q:", q)#SLOPE
```

```python
cost = (1/(2*len(x))) * sum(err**2)
print('Mean Squared error is: ',cost)
```

# Best Fit Line 

```python
y1  = q*x+p #PREDICTED
plt.scatter(x,y)
plt.plot(x,y1,"red")
```
